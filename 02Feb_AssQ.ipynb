*Q1 -> Explain with an example each when to use a for loop and a while loop.*

For loop is used when we know no. of iteration (i.e. no. of times a statement has to be executed) Example print table of 3 upto 10 times.

While loop is used when we know condition upto which loop has to terminated. Its used when no. of iteration is unknown. 

```python
# Print table using for loop 
for x in range(1,11,1):
  print(3*x, end = " ")
print()

# Print first 5 no divisible by 2 using while loop
i= 1
l1=[]
while i < 11:
  if i%2==0:
    l1.append(i)
  i= i+1
l1
```

* Q2-> Write a python program to print the sum and product of the first 10 natural numbers using for
and while loop.*

```python
# For loop
sum1 =0
Product1 =1
for i in range(1,11,1):
  sum1 = sum1 + i
  Product1 = Product1 * i

print("Sum of first 10 natural number", sum1)
print("Product of first 10 natural number", Product1)

# For loop
sum2 =0
Product2 =1
i=1
while i<11:
  sum2 = sum2 + i
  Product2 = Product2*i
  i =i+1

print("Sum of first 10 natural number", sum2)
print("Product of first 10 natural number", Product2)
```

Q3. Create a python program to compute the electricity bill for a household.

*The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit*

```python
unit = int(input('Enter units of electricity : '))
if unit < 100:
  print("Electricity bill :", unit*4.5)
elif unit < 200 and unit > 100:
  print("Electricity bill :", 100*4.5 + (unit-100)*6)  
elif unit < 300 and unit > 200:
  print("Electricity bill :", 100*4.5 + 100*6 + (unit-200)*10)
else :
  print("Electricity bill :", 100*4.5 + 100*6 + 100*10 + (unit-300)*20 )
```

Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print
that list.

```python

list1 = []
for i in range(1,101,1):
  if (i**3)%4 ==0 or (i**3)%5 ==0:
    list1.append(i)

print("Using for loop", list1)

i=1
list2 = []
while i<101:
  if (i**3)%4 == 0 or (i**3)%5 == 0:
    list2.append(i)
  i =i+1
print("Using while loop", list2)
```

Q5. Write a program to filter count vowels in the below-given string.

string = "I want to become a data scientist"

```python
list_vowels = ['a','e','i','o','u']
count=0
string = "I want to become a data scientist"
string = string.lower()
for i in range(len(string)):
  for j in range(len(list_vowels)):
    if string[i] == list_vowels[j] :
     count = count + 1
     i = i +1
     j = 0
     

print("Count of vowels in given string", count)
```
