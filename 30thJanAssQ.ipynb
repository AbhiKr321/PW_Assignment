```python
# Q6 use nested while loop to print 3 pattern in python

# Pattern 1 - Right triangle pattern:
print("Pattern 1 - Right triangle pattern:")
i = 1
while i <= 5:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1

# Pattern 2 - Square pattern:
print("Pattern 2 - Square pattern:")
i = 1
while i <= 5:
    j = 1
    while j <= 5:
        print("*", end="")
        j += 1
    print()
    i += 1

# Pattern 3 - Diamond pattern:
print("Pattern 3 - Diamond pattern:")
i = 1
while i <= 4:
    j = 1
    while j <= 4 - i:
        print(" ", end="")
        j += 1
    k = 1
    while k <= 2 * i - 1:
        print("*", end="")
        k += 1
    print()
    i += 1

i = 3
while i >= 1:
    j = 1
    while j <= 4 - i:
        print(" ", end="")
        j += 1
    k = 1
    while k <= 2 * i - 1:
        print("*", end="")
        k += 1
    print()
    i -= 1
```
